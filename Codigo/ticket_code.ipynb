```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
import datetime
from PIL import Image, ImageDraw, ImageFont

img = Image.open('ticket_formato.jpg')

fecha = datetime.datetime.now()
fecha.strftime('%d-%m-%Y')
hora = datetime.datetime.now()
hora.strftime('%H:%M')

reporte_falla = input('Por favor indique la opción deseada: ')

st_contador = 0
for ST in range(1000):
    
        if(reporte_falla != "1"):
            break
            
        if(reporte_falla == "1"):
            st_contador += 1
            
        if(st_contador <= 9):
            num_ticket = f"ST000{st_contador}"
            break
            
        elif(st_contador <= 99):
            num_ticket = f"ST00{st_contador}"
            break
        elif(st_contador <= 999): 
            num_ticket = f"ST0{st_contador}"
            break

if(reporte_falla == "1"):
    prioridad = "A"
    aprox = "10 min"

rt_contador = 0

for RT in range(1000):
    
    if(reporte_falla != "2"):
        break
        
    if(reporte_falla == "2"):
        rt_contador += 1 
        
    if(rt_contador <= 9):
        num_ticket = f"RT000{rt_contador}"
        break
    elif(rt_contador <= 99):
        num_ticket = f"RT00{rt_contador}"
        break
    elif(rt_contador <= 999):
        num_ticket = f"RT0{rt_contador}"
        break

if(reporte_falla == "2"):
    prioridad = "B"
    aprox = "20 min"   

dt_contador = 0

for DT in range(1000):
    
    if(reporte_falla != "3"):
        break
        
    if(reporte_falla == "3"):
        dt_contador += 1
        
    if(dt_contador <= 9):
        num_ticket = f"DT000{dt_contador}"
        break
    elif(dt_contador <= 99):
        num_ticket = f"DT00{dt_contador}"
        break
    elif(dt_contador <= 999):
        num_ticket = f"DT0{dt_contador}"
        break

if(reporte_falla == "3"):
    prioridad = "C"
    aprox = "30 min"
    

texto_hora = hora.strftime('%H:%M')    
texto_fecha = fecha.strftime('%d-%m-%Y')    
texto_ticket = num_ticket
texto_prioridad = prioridad
texto_aprox = aprox

img = Image.open('ticket_formato.jpg')
draw_fecha = ImageDraw.Draw(img)
draw_hora = ImageDraw.Draw(img)
draw_ticket = ImageDraw.Draw(img)
draw_prioridad = ImageDraw.Draw(img)
draw_aprox = ImageDraw.Draw(img)

font = ImageFont.truetype('calibri.ttf', 15)
font_ticket = ImageFont.truetype('calibri.ttf', 35)
font_prioridad = ImageFont.truetype('calibri.ttf', 30)
font_aprox = ImageFont.truetype('calibri.ttf', 20)

draw_fecha.text((25,90), texto_fecha, font=font, fill='black')
draw_hora.text((200,90), texto_hora, font=font, fill='black')
draw_ticket.text((80,245), texto_ticket, font=font_ticket, fill='black')
draw_prioridad.text((125,310), texto_prioridad, font=font_prioridad, fill='black')
draw_prioridad.text((110,355), texto_aprox, font=font_aprox, fill='black')

plt.imshow(img)
plt.show()
img.save(f"{num_ticket}.jpg")
```
